```python
def drawline(loc1,loc2,len1,len2,i):
    if i<num:
        p1=P1[i]
        p2=P2[i]
        if i%2:
            loc11=loc1
            loc12=[loc1[0]+len1[0]*p1,loc1[1]]
            len11=[len1[0]*p1,len1[1]]
            len12=[len1[0]*(1-p1),len1[1]]
            plt.plot([loc12[0],loc12[0]],[loc1[1],loc1[1]+len1[1]],color=colors[i])
            drawline(loc11,loc12,len11,len12,i+1)
            
            loc21=loc2
            loc22=[loc2[0]+len2[0]*p2,loc2[1]]
            len21=[len2[0]*p2,len2[1]]
            len22=[len2[0]*(1-p2),len2[1]]
            plt.plot([loc22[0],loc22[0]],[loc2[1],loc2[1]+len2[1]],color=colors[i])
            drawline(loc21,loc22,len21,len22,i+1)
        else:
            loc11=loc1
            loc12=[loc1[0],loc1[1]+len1[1]*p1]
            len11=[len1[0],len1[1]*p1]
            len12=[len1[0],len1[1]*(1-p1)]
            plt.plot([loc1[0],loc1[0]+len1[0]],[loc12[1],loc12[1]],color=colors[i])
            drawline(loc11,loc12,len11,len12,i+1)
            
            loc21=loc2
            loc22=[loc2[0],loc2[1]+len2[1]*p2]
            len21=[len2[0],len2[1]*p2]
            len22=[len2[0],len2[1]*(1-p2)]
            plt.plot([loc2[0],loc2[0]+len2[0]],[loc22[1],loc22[1]],color=colors[i])
            drawline(loc21,loc22,len21,len22,i+1)
            
    else:
        return

def drawchain():
    plt.plot([0,1,1,0,0],[0,0,1,1,0],color='k')
    plt.plot([0.5,0.5],[0,1],color='k')
    loc1=[0,0]
    loc2=[0.5,0]
    len1=[0.5,1]
    len2=[0.5,1]
    drawline(loc1,loc2,len1,len2,0)
    plt.axis('scaled')
    plt.axis('off')
```

```python
plt.rcParams['figure.figsize']=(15,int(Num/3*5))
for i in range(Num):
    num=i+1
    plt.subplot(int(Num/3),3,i+1)
    drawchain()

plt.show()
```
